**Data Understanding**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = 'https://raw.githubusercontent.com/Chelseaayu/Project_Chelsea/master/UserCarData.csv'
car = pd.read_csv(data)
car
```

```python
car.info()
```

```python
car.nunique()
```

**Data Preparation**

```python
hapus = ['seller_type', 'owner', 'seats', 'sold', 'torque']
car = car.drop(hapus, axis = 1)
car
```

```python
car['fuel'].unique()
```

```python
car['fuel'] = car['fuel'].replace({'Diesel' : 0, 'Petrol' : 1, 'LPG' : 2, 'CNG' : 3})
car
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x = car.drop('fuel', axis = 1)
y = car['fuel']

numeric_columns = ['max_power', 'mileage']
x_numeric = x[numeric_columns]

x_scaled = scaler.fit_transform(x_numeric)

x_scaled = pd.DataFrame(x_scaled, columns = numeric_columns)

print(x_scaled)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.3, random_state=42)
```

**Modelling**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
import xgboost as xgb
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# XGBoost
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
```

**Evaluation**

```python
actual = y_test  # Actual target values
# Create confusion matrices
knn_cm = confusion_matrix(actual, knn_pred)
dt_cm = confusion_matrix(actual, dt_pred)
rf_cm = confusion_matrix(actual, rf_pred)
xgb_cm = confusion_matrix(actual, xgb_pred)

# Convert confusion matrices to DataFrames
def confusion_matrix_to_dataframe(cm):
    labels = sorted(set(actual))
    df_cm = pd.DataFrame(cm, index=labels, columns=labels)
    df_cm.index.name = 'Actual'
    df_cm.columns.name = 'Predicted'
    return df_cm

# Convert confusion matrices to DataFrames with labels
knn_cm_df = confusion_matrix_to_dataframe(knn_cm)
dt_cm_df = confusion_matrix_to_dataframe(dt_cm)
rf_cm_df = confusion_matrix_to_dataframe(rf_cm)
xgb_cm_df = confusion_matrix_to_dataframe(xgb_cm)
```

```python
# Menghitung akurasi untuk model K-Nearest Neighbors (KNN)
knn_accuracy = accuracy_score(y_test, knn_pred)

# Menghitung akurasi untuk model Decision Tree
dt_accuracy = accuracy_score(y_test, dt_pred)

# Menghitung akurasi untuk model Random Forest
rf_accuracy = accuracy_score(y_test, rf_pred)

# Menghitung akurasi untuk model XGBoost
xgb_accuracy = accuracy_score(y_test, xgb_pred)

# Menampilkan hasil akurasi untuk setiap model
print("K-Nearest Neighbors (KNN) Accuracy:", knn_accuracy)
print("Decision Tree Accuracy:", dt_accuracy)
print("Random Forest Accuracy:", rf_accuracy)
print("XGBoost Accuracy:", xgb_accuracy)
```

```python
models = {
    'K-Nearest Neighbors (KNN)': knn_accuracy,
    'Decision Tree': dt_accuracy,
    'Random Forest': rf_accuracy,
    'XGBoost': xgb_accuracy
}

best_model_name = max(models, key=models.get)
print("Best Model:", best_model_name)
```

```python
car_new = pd.read_csv('banyak_mobil.csv')
car_new
```

```python
# Filter data dengan nilai 'Jenis Bahan Bakar' (fuel) sama dengan 3
filtered_result = car_new[car_new['fuel'] == 3]

# Kemudian, mengelompokkan berdasarkan 'City' dan menghitung jumlahnya
city_counts = filtered_result.groupby('City')['Jumlah Mobil'].sum()

# mengurutkan dari yang tertinggi
city_counts_sort = city_counts.sort_values(ascending = False)

# Menampilkan jumlah mobil dari masing-masing kota dengan 'Jenis Bahan Bakar' = 3
print(city_counts_sort)
```

```python
# Filter data dengan nilai 'Jenis Bahan Bakar' (fuel) sama dengan 2
lpg_car = car_new[car_new['fuel'] == 2]

# Kemudian, mengelompokkan berdasarkan 'City' dan menghitung jumlahnya
lpg_car_count = lpg_car.groupby('City')['Jumlah Mobil'].sum()

# mengurutkan dari yang tertinggi
lpg_counts_sort = lpg_car_count.sort_values(ascending = False)

# Menampilkan jumlah mobil dari masing-masing kota dengan 'Jenis Bahan Bakar' = 2
print(lpg_counts_sort)
```

Dari 2 data diatas, dimana data pertama adalah kota yang memiliki mobil berbahan bakar CNG ( Gas Alam ) dan data kedua adalah kota yang memiliki mobil berbahan bakar LPG. Dilansir dari [Cummins Newsroom](https://www.cummins.com/news/2022/05/05/natural-gas-cng-vs-lpg-lng-rng-and-diesel) bahwasanya bahan bakar yang bisa dibilang ramah lingkungan adalah bahan bakar LPG dan CNG. Dengan begitu dapat dilihat bahwasanya pentingnya masyarakat ikut berkontribusi dalam memilih mobil ramah lingkungan mayoritas sudah dimengerti, akan tetapi baru sedikit. Oleh karena itu, diharapkan kepada kota-kota lain untuk lebih menghimbau masyarakatnya untuk menggunakan kendaraan berbahan bakar ramah lingkungan

```python
# Filter data dengan nilai 'Jenis Bahan Bakar' (fuel) sama dengan 3 dan 2
lpg_car = car_new[car_new['fuel'] == 2]
cng_car = car_new[car_new['fuel'] == 3]

# Kemudian, mengelompokkan berdasarkan 'Name' dan menghitung jumlahnya
lpg_car_count = lpg_car.groupby('name')['Jumlah Mobil'].sum()
cng_car_count = cng_car.groupby('name')['Jumlah Mobil'].sum()

# mengurutkan dari yang tertinggi
lpg_counts_sort = lpg_car_count.sort_values(ascending = False)
cng_counts_sort = cng_car_count.sort_values(ascending = False)

# Menampilkan jumlah mobil dari masing-masing kota dengan 'Jenis Bahan Bakar' = 3
print('Jumlah Mobil Berbahan Bakar LPG dari masing-masing merk :', lpg_counts_sort)
print('Jumlah Mobil Berbahan Bakar CNG dari masing-masing merk :', cng_counts_sort)
```

Dari data diatas dapat kita lihat bahwasanya masih sedikit merk mobil yang memiliki kendaraan berbahan bakar ramah lingkungan. Harapannya semoga merk mobil lainnya dapat segera meluncurkan mobil ramah lingkungannya
