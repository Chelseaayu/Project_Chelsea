```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read dataset
df = pd.read_csv('Pokemon.csv')
df
```

```python
# describe dataset
df.info()
```

```python
# unique test
df.nunique()
```

```python
# change column name
df = df.rename(columns={'Type 1' : 'type_1', 'Type 2' : 'type_2', 'Sp. Atk' : 'sp_atk', 'Sp. Def' : 'sp_def'})
df
```

```python
# delete column #
df = df.drop(['#'], axis=1)
df
```

```python
# missing value
df.isnull().sum()
```

```python
# Count the number of Pokemon for each type
type_1_counts = df['type_1'].value_counts()
type_2_counts = df['type_2'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(type_1_counts.index, type_1_counts.values, label='First Type', alpha=0.7)
plt.bar(type_2_counts.index, type_2_counts.values, label='Second Type', alpha=0.7)
plt.xlabel('Pokemon Type')
plt.ylabel('Number of Pokemon')
plt.title('Number of Pokemon by Type')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# hitung banyaknya masing-masing tipe pada type_1
type_1_counts = df['type_1'].value_counts()
print(type_1_counts)
```

```python
# hitung banyaknya masing-masing tipe pada type_2
type_2_counts = df['type_2'].value_counts()
print(type_2_counts)
```

```python
# Ambil baris yang type_2 nya NaN
nan_type_2 = df[df['type_2'].isna()]

# Visualisasikan berdasarkan legendary
plt.figure(figsize=(8, 6))
sns.countplot(data=nan_type_2, x='Legendary')
plt.xlabel('Legendary')
plt.ylabel('Sum')
plt.title('Pokemon that does not have a secondary type based on its legendary')

# Tambahkan jumlah di atas setiap bar
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                        textcoords='offset points')

plt.tight_layout()
plt.show()

# Tuliskan jumlahnya
false_count = nan_type_2['Legendary'].value_counts()[False]
true_count = nan_type_2['Legendary'].value_counts()[True]

print(f"Not Legendary: {false_count}")
print(f"Legendary: {true_count}")
```

```python
# Ambil baris yang type_2 nya tidak NaN
not_nan_type_2 = df[~df['type_2'].isna()]

# Visualisasikan berdasarkan legendary
plt.figure(figsize=(8, 6))
sns.countplot(data=not_nan_type_2, x='Legendary')
plt.xlabel('Legendary')
plt.ylabel('Sum')
plt.title('Pokemon that has a secondary type based on its legendary')

# Tambahkan jumlah di atas setiap bar
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                        textcoords='offset points')

plt.tight_layout()
plt.show()

# Tuliskan jumlahnya
false_count = not_nan_type_2['Legendary'].value_counts()[False]
true_count = not_nan_type_2['Legendary'].value_counts()[True]

print(f"Not Legendary: {false_count}")
print(f"Legendary: {true_count}")
```

```python
# encoding type 1
encod_type1 = 'type_1'

unique_values_type1 = df[encod_type1].unique()
encoding_mapping_type1 = {value: index for index, value in enumerate(unique_values_type1)}
df['type1_encoding'] = df[encod_type1].map(encoding_mapping_type1)
df
```

```python
encoding_mapping_type1
```

```python
# prompt: hapus baris yang type 2 nya nan

df = df.dropna(subset=['type_2'])
df
```

```python
# encoding type 2
encod_type2 = 'type_2'

unique_values_type2 = df[encod_type2].unique()
encoding_mapping_type2 = {value: index for index, value in enumerate(unique_values_type2)}
df['type2_encoding'] = df[encod_type2].map(encoding_mapping_type2)
df
```

```python
encoding_mapping_type2
```

```python
# Select columns for correlation, excluding 'Name', 'type_1', and 'type_2'
selected_columns = df.drop(['Name', 'type_1', 'type_2'], axis=1)

# Calculate correlation matrix
correlation_matrix = selected_columns.corr()

# Create heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Correlation')
plt.show()
```

```python
# Hitung banyaknya masing-masing tipe pada type_1
type_1_counts = df['type_1'].value_counts()

# Buat bar plot dengan warna yang berbeda
plt.figure(figsize=(10, 6))
colors = ['#1F77B4', '#FF7F0E', '#2CA02C', '#D62728', '#9467BD', '#8C564B', '#E377C2', '#7F7F7F', '#BCBD22', '#17BECF', '#F5B041']
bars = type_1_counts.plot(kind='bar', color=colors[:len(type_1_counts)])

# Tambahkan anotasi untuk menampilkan jumlah di atas setiap batang
for p in bars.patches:
    bars.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                  textcoords='offset points')

plt.xlabel('First Type')
plt.ylabel('Count')
plt.title('Number of Pokemon by First Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Hitung banyaknya masing-masing tipe pada type_1
type_2_counts = df['type_2'].value_counts()

# Buat bar plot dengan warna yang berbeda
plt.figure(figsize=(10, 6))
colors = ['#1F77B4', '#FF7F0E', '#2CA02C', '#D62728', '#9467BD', '#8C564B', '#E377C2', '#7F7F7F', '#BCBD22', '#17BECF', '#F5B041']
bars = type_2_counts.plot(kind='bar', color=colors[:len(type_2_counts)])

# Tambahkan anotasi untuk menampilkan jumlah di atas setiap batang
for p in bars.patches:
    bars.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                  textcoords='offset points')

plt.xlabel('Secondary Type')
plt.ylabel('Count')
plt.title('Number of Pokemon by Secondary Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# prompt: visualisasikan hubungan total dengan legendary. yang diinginkan adalah dikatakan legendary itu jika totalnya banyak atau tidak

# Buat scatter plot untuk melihat hubungan antara Total dan Legendary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Total', y='Legendary', hue='Legendary', s=100)
plt.xlabel('Total')
plt.ylabel('Legendary')
plt.title('Relationship between Total and Legendary')
plt.show()

# Hitung rata-rata Total untuk Pokemon Legendary dan non-Legendary
legendary_mean = df[df['Legendary'] == True]['Total'].mean()
non_legendary_mean = df[df['Legendary'] == False]['Total'].mean()

print(f"Rata-rata Total untuk Pokemon Legendary: {legendary_mean:.2f}")
print(f"Rata-rata Total untuk Pokemon non-Legendary: {non_legendary_mean:.2f}")

# Buat boxplot untuk visualisasi yang lebih jelas
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Legendary', y='Total')
plt.xlabel('Legendary')
plt.ylabel('Total')
plt.title('Distribution of Total for Legendary and non-Legendary Pokemon')
plt.show()

# Kesimpulan: Dari visualisasi dan perhitungan rata-rata, terlihat bahwa Pokemon Legendary cenderung memiliki Total yang lebih tinggi dibandingkan dengan Pokemon non-Legendary.
```

```python
# Buat point plot untuk visualisasi yang lebih jelas
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x='Legendary', y='Speed', capsize=0.1)
plt.xlabel('Legendary')
plt.ylabel('Speed')
plt.title('Distribution Speed for Legendary and non-Legendary Pokemon with Confidence Intervals')
plt.show()
```

```python
# Buat histogram untuk visualisasi yang lebih jelas
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='sp_atk', hue='Legendary', multiple='stack', bins=20)
plt.xlabel('Special Attack')
plt.ylabel('Legendary')
plt.title('Histogram of Special Attack for Legendary and non-Legendary Pokemon')
plt.show()
```

```python
# Buat KDE plot untuk visualisasi yang lebih jelas
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='sp_def', hue='Legendary', fill=True)
plt.xlabel('Special Defense')
plt.ylabel('Density')
plt.title('KDE Plot of Special Defense for Legendary and non-Legendary Pokemon')
plt.show()
```
